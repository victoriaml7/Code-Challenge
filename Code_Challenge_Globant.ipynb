```python
import pandas as pd
```

Cargamos el archivo csv

```python
df = pd.read_csv('code.csv', sep=',')
```

```python
df
```

Asignamos transaction_id como index

```python
df.set_index('transaction_id', inplace=True)
```

Visualizamos

```python
df.head(5)
```

```python
df.info()
```

Revisamos los nulos

```python
df.isnull()
```

Borramos el duplicado

```python
df.dropna(subset = ['price'], inplace = True)
```

```python
df[df['price'].isnull()]
```

Podemos ver que quantity_sold (cantidad vendida en español), es de tipo float. Esto ocurre porque el campo es nulo.
Podemos cambiar el tipo de dato ya que, el tipo float admite números negativos, positivos y decimales. Por el tipo de articulos vendidos, este tipo de dato no es lo ideal porque no puedo vender 1.5 Laptops, por lo que es recomendado cambiarlo a INT

```python
df['quantity_sold'] = df.quantity_sold.astype('int')
df.info()
```

```python
df
```

Podemos ver que user_agent tiene un dato faltante

```python
df[['user_agent']]
```

Se consultó al equipo y se puede llenar reemplazar el campo "" de user_agent con “Mozilla/5.0 (Android)”

Realizamos el cambio

```python
df['user_agent'] = df['user_agent'].str.replace('“”', '“Mozilla/5.0 (Android)”',regex=True)
df
```

date_sold debería ser tipo date, sin embargo es object, podemos ver que hay un espacio extra después del segundo -, lo corregimos

```python
df['date_sold'] = df['date_sold'].str.replace(' ','',regex=True)
```

```python
df
```

```python
df['date_sold'] = pd.to_datetime(df['date_sold'])
```

```python
df.info()
```
